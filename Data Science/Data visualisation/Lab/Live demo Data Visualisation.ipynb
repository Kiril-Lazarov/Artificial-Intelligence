```python
# Plot histogram
plt.hist(report_2018['GDP per capita'])
plt.xlabel('GDP contribution')
plt.ylabel('Count')
plt.show()
```

```python
report_2018['GDP per capita'].mean()
```

```python
# Missing data
report_2018[report_2018['GDP per capita'] > 1.5].sort_values('GDP per capita')
```

```python
report_2018.nsmallest(10, 'GDP per capita')
```

```python
plt.hist(report_2018.nsmallest(20, 'GDP per capita')['GDP per capita'])
plt.show()
```

```python
report_2019[report_2019['GDP per capita'] > 1.5].sort_values('GDP per capita')
```

```python
# Plot two datasets 
plt.hist(report_2018['GDP per capita'], label = '2018', alpha = 0.7)
plt.hist(report_2019['GDP per capita'], label = '2019', alpha = 0.7)
plt.legend()
plt.xlabel('GDP contribution')
plt.ylabel('Count')
plt.show()
```

```python
# Plot two datasets for score
plt.hist(report_2018['Score'], label = '2018', alpha = 0.7)
plt.hist(report_2019['Score'], label = '2019', alpha = 0.7)
plt.legend()
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot
plt.boxplot([report_2018['Score'],report_2019['Score']])
plt.xticks((1,2), ['2018', '2019'])
plt.show()
```

```python
# Merging two datasets
report_2018.merge(report_2019, left_on = 'Country or region', right_on = 'Country or region')
```

```python
# Replace x and y suffixes
merged_reports = report_2018.merge(report_2019, left_on = 'Country or region', right_on = 'Country or region', suffixes = ['_2018', '_2019'] )
```

```python
merged_reports
```

```python
scores = merged_reports[['Country or region', 'Score_2018', 'Score_2019']]
```

```python
scores
```

```python
scores.nlargest(20, 'Score_2018')
```

```python
# Bar charts are used to plot categorical variables
plt.barh(scores.nlargest(20, 'Score_2018')['Country or region'], scores.nlargest(20, 'Score_2018')['Score_2018'])
plt.gca().invert_yaxis()
plt.show()
```

```python
# Bar charts are used to plot categorical variables
plt.barh(scores.nsmallest(20, 'Score_2018')['Country or region'], scores.nsmallest(20, 'Score_2018')['Score_2018'])
plt.show()
```
