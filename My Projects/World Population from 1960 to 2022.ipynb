```python
def show_hist(country_code):
    data = np.array(world_population.loc[world_population['Country Code'] == country_code].drop(columns = ['Country Name', 'Country Code']))[0]
    mean_population = np.mean(data)

    # plt.figure(figsize=(6, 6))
    plt.hist(data, bins=15, edgecolor='black')
    max_height = np.max(data)
    mean_x = np.mean(data)      
    print(np.max(data), mean_x,max_height)
    
    plt.title('Population Histogram of {}'.format(country_code))

    plt.xlabel('Population')
    plt.ylabel('Count')
    
    plt.text(mean_x, max_height, f'Mean population between 1960-2022 - {mean_population:.0f}', fontsize=9,ha = 'center')
    
    plt.show()
show_hist('BGR')
```

```python
def show_hist(country_code):
    data = np.array(world_population.loc[world_population['Country Code'] == country_code].drop(columns=['Country Name', 'Country Code']))[0]
    mean_population = np.mean(data)

    plt.figure(figsize=(10, 6))  # Променете размера на фигурата за да вмести пълния текст
    max_height= plt.hist(data, bins=15, edgecolor='black')

    mean_x = np.mean(data)
    
    plt.title('Population Histogram of {}'.format(country_code))

    plt.xlabel('Population')
    plt.ylabel('Count')

    plt.text(mean_x, max_height[0].max(), f'Mean population between 1960-2022 - {mean_population:.0f}', fontsize=9, ha='center')

    plt.show()

show_hist('USA')
```
