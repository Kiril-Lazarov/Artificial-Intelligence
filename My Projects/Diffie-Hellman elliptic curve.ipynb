```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root

# Дефиниране на спиралата
t = np.linspace(0, 100, 1000)
speed = 1
angle = np.pi / 32
x_axis = t * speed * np.cos(t * angle)
y_axis = t * speed * np.sin(t * angle)

# Дефиниране на отсечката
A = np.array([50, -50])
B = np.array([-25, 25])
slope = (B[1] - A[1]) / (B[0] - A[0])
const = slope * A[0] - A[1]

# Функция, която връща вектора от стойности на системата от уравнения
def equations(vars):
    x, y = vars
    eq1 = x - t * speed * np.cos(t * angle)
    eq2 = y - t * speed * np.sin(t * angle) - (slope * x + const)
    return [eq1, eq2]

# Решаване на системата от уравнения
sol = root(equations, [0, 0])
t_intersections = sol.x

# Изобразяване на графиката
plt.plot(x_axis, y_axis)
plt.plot([A[0], B[0]], [A[1], B[1]])
plt.scatter(A[0], A[1], color='blue', label='Точка A')
plt.scatter(B[0], B[1], color='green', label='Точка B')
plt.scatter(t_intersections * speed * np.cos(t_intersections * angle),
            t_intersections * speed * np.sin(t_intersections * angle),
            color='red', label='Пресечни точки')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Спирала и отсечка')
plt.grid(True)
plt.show()

print("Пресечни точки:")
for i, t_val in enumerate(t_intersections):
    x_val = t_val * speed * np.cos(t_val * angle)
    y_val = t_val * speed * np.sin(t_val * angle)
    print(f"Точка {i+1}: ({x_val}, {y_val})")
```

```python
def find_nth_derivative(expr,nth,center):
    return diff(expr,x,nth).evalf(subs = {x:center})


def taylor_coeffs(function,nth,center):
    result = []
    for n in range(nth):
      
        deriv = find_nth_derivative(function,n,center)
        # taylor_coefficient = (deriv/math.factorial(n))
        result.append(deriv)
    return result

def calc_polynomial(polynomial_coefficients, x_value,center):
    result = 0
    for index in range(len(polynomial_coefficients)):
        result+= (polynomial_coefficients[index] * ((x_value-center) ** index))/math.factorial(index)
    return result


def plot_taylor_series(function,derivatives_count,center,x_min=-1,x_max=1):
    
    x_axis = np.linspace(x_min,x_max,500)
    y_original = [find_nth_derivative(function,0,x) for x in x_axis]
    plt.title('Taylor series plots')
    plt.plot(x_axis, y_original, label = r'Original function')
    plt.scatter(center,find_nth_derivative(function,0,center))
    ax = plt.gca()
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.xlabel('x')
    plt.ylabel('y')

    for nth_derivative in range(derivatives_count):
        nth_dy = [calc_polynomial(taylor_coeffs(function,nth_derivative + 2,center), x, center) for x in x_axis]
        plt.plot(x_axis,nth_dy, label = f'{nth_derivative + 1} derivative')
    plt.legend()
    plt.show()
```

```python
import pygame
import numpy as np
def animation(screen_width = 1500,screen_height = 800):   
    
    

  

    center = (screen_width / 2, screen_height / 2)
    
    pygame.init()
    win = pygame.display.set_mode((screen_width, screen_height))
    bg_color = (255, 255, 255)

    pygame.display.set_caption('Expanding spiral')
    
   
    angle_speed = np.pi/4
    font = pygame.font.Font(None, 36)

    clock = pygame.time.Clock()
    FPS = 100  # frames per second
    clock.tick(FPS)

    running = True
    frame = 0

    
        
    while running:
        frame += 1
        speed = frame/4
        # Check for event
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        win.fill(bg_color)
        # position = np.array(center)

        # Draw expanding cirlce
        pygame.draw.circle(win, (255, 165, 0), center, speed)
        pygame.draw.circle(win, (255, 255, 255), center, speed-1)
        
        # Draw axeses
        pygame.draw.line(win, (0,0,0), [750,0], [750,800], 1)
        pygame.draw.line(win, (0,0,0), [0,400], [1500,400], 1)        
        
        x_position = speed * np.cos(angle_speed * speed/10) + center[0]
        y_position = speed * np.sin(angle_speed * speed/10) + center[1]
        pygame.draw.circle(win, (0, 0, 0), [x_position, y_position], 3)
        
        pygame.draw.line(win, (250,0,0),[x_position,center[1]], [center[0],y_position],2)
    
        
        text = font.render(f'Frame number {frame}', True, (0, 0, 0))
        # Refreshing the screen
        win.blit(text, (10, 10))
        clock.tick(FPS)
        pygame.display.update()
        pygame.display.flip()
        
        if frame == 2000:
            running = False
       
    # Exit
    pygame.quit()
animation()
```
